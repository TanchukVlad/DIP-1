```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
#Доп задания 
chess=np.zeros((250,250),dtype=np.uint8)
k=1                              # переключатель
for i in range(len(chess)):
  if (i%50==0):
    k=~k+2                        # переключается с 1 в 0 и на оборот через каждых 50 строк 
  for N in range(k,6,2):          #делаем белые квадраты
    chess[i][50*N:50*(N+1)]=255 
  plt.imshow(chess,cmap="gray")
```

```
#Laba задания 
img_half = cv2.imread("assignments/Lab3_half100.png")
img_chess = cv2.imread("assignments/Lab3_chess100.png")
hist_half = cv2.calcHist([img_half],[0],None,[256],[0, 256])
hist_chess = cv2.calcHist([img_chess],[0],None,[256],[0, 256])

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_half,cmap="gray")
plt.subplot(122)
plt.plot(hist_half); plt.grid()
plt.show()

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_chess,cmap="gray")
plt.subplot(122)
plt.plot(hist_chess); plt.grid()
plt.show()
```
