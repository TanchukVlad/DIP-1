1.1. Відкрити та завантажити наступні два зображення:
    * assignments/Lab3_half100.png
    * assignments/Lab3_chess100.png

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_half = cv2.imread("assignments/Lab3_half100.png")
img_chess = cv2.imread("assignments/Lab3_chess100.png")
```

1.2.Визначити гістограму для кожного зображення:

```python
#img_half
hist_half = cv2.calcHist([img_half],[0],None,[256],[0, 256])
#img_chess
hist_chess = cv2.calcHist([img_chess],[0],None,[256],[0, 256])
```

1.3.Відобразити завантаженні зображення та відповідно визначені гістограми зображень за допомогою функції **plt.subplot**:

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_half,cmap="gray")
plt.subplot(122)
plt.plot(hist_half); plt.grid()
plt.show()

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_chess,cmap="gray")
plt.subplot(122)
plt.plot(hist_chess); plt.grid()
plt.show()
```

1.4 Виконати просторову фільтрацію зображень використовуючи згладжувальний фільтра або розмивання Гауса з  розмірами - 3x3, 5x5, 9x9, 15x15.

```python
half_blur3 = cv2.GaussianBlur(img_half, (3,3), 0)
half_blur5 = cv2.GaussianBlur(img_half, (5,5), 0)
half_blur9 = cv2.GaussianBlur(img_half, (9,9), 0)
half_blur15 = cv2.GaussianBlur(img_half, (15,15), 0)

chess_blur3 = cv2.GaussianBlur(img_chess, (3,3), 0)
chess_blur5 = cv2.GaussianBlur(img_chess, (5,5), 0)
chess_blur9 = cv2.GaussianBlur(img_chess, (9,9), 0)
chess_blur15 = cv2.GaussianBlur(img_chess, (15,15), 0)
```

1.5 Визначити гістограми для всіх отриманих зображень.

```python
half_blur3_hist = cv2.calcHist([half_blur3],[0],None,[256],[0, 256])
half_blur5_hist = cv2.calcHist([half_blur5],[0],None,[256],[0, 256])
half_blur9_hist = cv2.calcHist([half_blur9],[0],None,[256],[0, 256])
half_blur15_hist = cv2.calcHist([half_blur15],[0],None,[256],[0, 256])

chess_blur3_hist = cv2.calcHist([chess_blur3],[0],None,[256],[0, 256])
chess_blur5_hist = cv2.calcHist([chess_blur5],[0],None,[256],[0, 256])
chess_blur9_hist = cv2.calcHist([chess_blur9],[0],None,[256],[0, 256])
chess_blur15_hist = cv2.calcHist([chess_blur15],[0],None,[256],[0, 256])
```

1.6.Відобразити отримані після фільтрації зображення та відповідно визначені гістограми за допомогою функції plt.subplot:

```python
plt.figure(figsize=(20,10))
plt.subplot(421)
plt.imshow(half_blur3,cmap="gray")
plt.subplot(422)
plt.plot(half_blur3_hist); plt.grid()
plt.subplot(423)
plt.imshow(half_blur5,cmap="gray")
plt.subplot(424)
plt.plot(half_blur5_hist); plt.grid()
plt.subplot(425)
plt.imshow(half_blur9,cmap="gray")
plt.subplot(426)
plt.plot(half_blur9_hist); plt.grid()
plt.subplot(427)
plt.imshow(half_blur15,cmap="gray")
plt.subplot(428)
plt.plot(half_blur15_hist); plt.grid()

plt.figure(figsize=(20,10))
plt.subplot(421)
plt.imshow(chess_blur3,cmap="gray")
plt.subplot(422)
plt.plot(chess_blur3_hist); plt.grid()
plt.subplot(423)
plt.imshow(chess_blur5,cmap="gray")
plt.subplot(424)
plt.plot(chess_blur5_hist); plt.grid()
plt.subplot(425)
plt.imshow(chess_blur9,cmap="gray")
plt.subplot(426)
plt.plot(chess_blur9_hist); plt.grid()
plt.subplot(427)
plt.imshow(chess_blur15,cmap="gray")
plt.subplot(428)
plt.plot(chess_blur15_hist); plt.grid()

plt.show()
```

Extra task:
Фільтри підвищення різкості

```python
img = cv2.imread('./examples/Fig0340a.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray') 
plt.axis('off')
```

```python
img=img.astype(float)
filter_hor = np.array([[-1, -2, -1], 
                       [0, 0, 0], 
                       [1, 2, 1]])
filter_ver= np.array([[-1, 0, 1], 
                      [-2, 0, 2], 
                      [-1, 0, 1]])

img_hor=cv2.filter2D(img, -1, filter_hor)
img_ver=cv2.filter2D(img, -1, filter_ver)

img_hor=cv2.convertScaleAbs(img_hor)
img_ver=cv2.convertScaleAbs(img_ver)

plt.figure(figsize=(30, 10))
plt.subplot(131)
plt.imshow(img, cmap='gray'), plt.title('Original'), plt.axis('off')
plt.subplot(132)
plt.imshow(img_hor, cmap='gray'), plt.title('horizontal'), plt.axis('off')
plt.subplot(133)
plt.imshow(img_ver, cmap='gray'), plt.title('vertical'), plt.axis('off')
plt.show()
```

```python
img_add=cv2.addWeighted(img_hor,0.5,img_ver,0.5,0)
plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(img, cmap='gray'), plt.title('Original'), plt.axis('off')
plt.subplot(122)
plt.imshow(img_add, cmap='gray'), plt.title('Borders filter'), plt.axis('off')
plt.show()
```
